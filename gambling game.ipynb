## 抛硬币游戏

```python
'''
抛硬币游戏：
抛硬币，正反两面概率各50%，正面你赢，反面庄家赢，每次下注1元，赔率1:1。
10位玩家，赌本10元，进行100轮。
'''
sample_list = []
round_num = 100
player_num = 10
for player in range(1, player_num + 1):
    gamble_fund = 10
    for r in range(1, round_num + 1):
        coin = random.randint(0,1)   # 0为正，1为负
        if coin == 0:
            gamble_fund += 1
        if coin == 1:
            gamble_fund -= 1
        if gamble_fund == 0:
            sample_list.append([player, r, gamble_fund])
            break
        else:
            pass
    sample_list.append([player, r, gamble_fund])
sample_1 = pd.DataFrame(sample_list,columns=['player','round','gamble_fund'])
sample_1
```

## 画图: cumulated # of players by # of rounds

```python
#默认像素：[6.0, 4.0]，默认分辨率：100，图片尺寸为 600*400
#指定 dpi=200，图片尺寸为 1200*800
#指定 dpi=300，图片尺寸为 1800*1200
#设置figsize可以在不改变分辨率的情况下改变比例
mpl.rcParams['savefig.dpi'] = 100  #图片像素
mpl.rcParams['figure.dpi'] = 300  #分辨率
mpl.rcParams['figure.figsize'] = (5.0, 4.0)   #显示图像的最大范围
mpl.rcParams['image.interpolation'] = 'nearest'  #插值方式
mpl.rcParams['image.cmap'] = 'gray'    #空间灰度

mpl.rcParams['font.sans-serif'] = 'SimHei'  #设置字体为SimHei显示中文
mpl.rcParams['axes.unicode_minus'] = False  #设置正常显示字符
mpl.rcParams['lines.linestyle'] = '-.'  #设置线条样式
mpl.rcParams['lines.linewidth'] = 3  #设置线条宽度
```

```python
sample3_data = sample_3.groupby(['r']).size().cumsum().reset_index(name='cumulated # players')
sample3_data
```

```python
# use pandas.plot to plot graph
sample3_data.plot(
       title = 'Game Over Round',
       kind = 'line',
       linewidth = 7,
       x = 'r',
       y = 'cumulated # players',
       xlim = [0,3500],    #设置x坐标轴范围
       figsize = (12,6),
       style = '-',
       color = 'gray',
       legend = True, 
       alpha = 0.8,   #alpha设置透明度
       rot = 45,   #设置x标签的旋转角度
       grid = True)

plt.xlabel('round')

# 在图上加标注
## eg: plt.axvline(x = 2.5, ymin = 0.5, ymax = 1, color = 'red)
plt.axvline(sample3_data['r'].median(),color='red')   #在坐标轴上添加垂直线
plt.axhline(50, color='red')      #在坐标轴上添加水平线
## plt.text()用于设置文字说明
## plt.text(x,y,string,fontsize=15, verticalalignment='top',horizontalalignment='right')
plt.text(300,60,'bankrupcy median on the {} round'.format(int(sample3_data['r'].median())), color='blue')
plt.text(200, 45,'half of the players bankrupty on the {} round'.format(sample3_data[sample3_data['cumulated # players']== 50]['r'].astype(int)),color='blue')
plt.savefig('./十元赌本无限轮结果图.png', dpi=400)
plt.show()
```

## 对赌游戏

```python
'''
抛硬币游戏：
抛硬币，正反两面概率各50%，正面你赢，反面庄家赢，每次下注1元，赔率1:1。
1000位玩家，赌本10元，庄家赌本10元，对赌，直到一方破产。
'''
sample_list = []
player_num = 1000
for player in range(1, player_num+1):
    dealer_gamble_fund = 10
    player_gamble_fund = 10
    r = 0
    while 1:   # = while True，一般用于想连续执行某个你不知道次数，只知道终止条件的操作
        r += 1
        coin = random.randint(0,1)   # 0为正，1为负
        if coin == 0:
            dealer_gamble_fund -= 1
            player_gamble_fund += 1
        if coin == 1:
            dealer_gamble_fund += 1
            player_gamble_fund -= 1
        if dealer_gamble_fund == 0 or player_gamble_fund == 0:
            sample_list. append([player, player_gamble_fund, dealer_gamble_fund, r])
            break
    # print('第{}位玩家进行了{}轮游戏'.format(player,r))
sample_4 = pd.DataFrame(sample_list, columns=['player','player_fund','dealer_fund','round'])
sample_4.to_csv('./对赌各十元赌本直到破产.csv', index=0)

print('玩家赢了{}次，赢钱概率为{}'.format(sample_4[sample_4['player_fund']>0]['player'].count(),'%s%%'%(sample_4[sample_4['player_fund']>0]['player'].count()/1000*100)))
print('庄家赢了{}次，赢钱概率为{}'.format(sample_4[sample_4['dealer_fund']>0]['player'].count(),'%s%%'%(sample_4[sample_4['dealer_fund']>0]['player'].count()/1000*100)))
# 对赌双方赢的概率均约为50%
```
