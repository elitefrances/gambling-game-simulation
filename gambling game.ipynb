```python
import random
import pandas as pd
import time
```

```python
'''
抛硬币游戏：
抛硬币，正反两面概率各50%，正面你赢，反面庄家赢，每次下注1元，赔率1:1。
10位玩家，赌本10元，进行100轮。
'''
sample_list = []
round_num = 100
player_num = 10
for player in range(1, player_num + 1):
    gamble_fund = 10
    for r in range(1, round_num + 1):
        coin = random.randint(0,1)   # 0为正，1为负
        if coin == 0:
            gamble_fund += 1
        if coin == 1:
            gamble_fund -= 1
        if gamble_fund == 0:
            sample_list.append([player, r, gamble_fund])
            break
        else:
            pass
    sample_list.append([player, r, gamble_fund])
sample_1 = pd.DataFrame(sample_list,columns=['player','round','gamble_fund'])
sample_1
```

```python
'''
抛硬币游戏：
抛硬币，正反两面概率各50%，正面你赢，反面庄家赢，每次下注1元，赔率1:1。
100位玩家，赌本10元，进行1000轮。
'''
sample_list = []
round_num = 1000
player_num = 100
for player in range(1, player_num + 1):
    gamble_fund = 10
    for r in range(1, round_num + 1):
        coin = random.randint(0,1)   # 0为正，1为负
        if coin == 0:
            gamble_fund += 1
        if coin == 1:
            gamble_fund -= 1
        if gamble_fund == 0:
            sample_list.append([player, r, gamble_fund])
            break
        else:
            pass
    sample_list.append([player, r, gamble_fund])
sample_2 = pd.DataFrame(sample_list, columns=['player', 'r', 'gamble_fund']).sort_values('gamble_fund', ascending=False).drop_duplicates()
sample_2.to_csv('./十元赌本1000轮游戏.csv',index=0)
```
